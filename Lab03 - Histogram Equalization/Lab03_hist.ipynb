```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("im_cells.png", cv2.IMREAD_GRAYSCALE)

imgOut = 256 - img

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,2,2)
plt.title('Invertida')
plt.imshow(imgOut, cmap='gray', vmin=0, vmax=255)
```

- Abra a imagem 'pout.tif' e mostre seu histograma.
- Equalize a imagem anterior e mostre novamente seu histograma. Mostra a curva de transformação (consulte https://www.mathworks.com/help/images/histogram-equalization.html)
(procurar função OpenCV)



```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

imgOriginal  = cv2.imread('pout.tif', cv2.IMREAD_GRAYSCALE)
plt.subplot(1,2,1)
plt.title("Imagem Original")
plt.hist(imgOriginal)

imgEqualized = cv2.equalizeHist(imgOriginal)
plt.subplot(1,2,2)
plt.title("Imagem Equalizada")
plt.hist(imgEqualized)
```
